```python
import pandas as pd

# Path to your CSV file
path_train = 'data/train_data.csv'
path_test = 'data/test_data.csv'
path_test_label = 'data/test_labels.csv'

# Read the CSV file into a DataFrame
df_train = pd.read_csv(path_train)
df_test = pd.read_csv(path_test)
df_test_labels = pd.read_csv(path_test_label)

# No missing values
# [0,1] => Distribution ist nach normalisierung gleich
```

```python
df_train
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math
# Number of columns
num_columns = len(df_train.columns)

# Calculate the number of rows and columns for the subplots
num_rows = math.ceil(num_columns / 3)
num_cols = min(num_columns, 3)

# Create subplots
fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, num_rows * 5))

# Flatten the axes array if there's more than one plot
axes = axes.flatten() if num_columns > 1 else [axes]

# Plot data distribution for each column
for i, column in enumerate(df_train.columns):
    axes[i].hist(df_train[column], bins=10, edgecolor='black', alpha=0.7)
    axes[i].set_title(f'Distribution of {column}')
    axes[i].set_xlabel(column)
    axes[i].set_ylabel('Frequency')

# Remove any unused axes
for ax in axes[num_columns:]:
    ax.remove()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def scatter(data):
    x = data[:, 0]  # First column for x-axis
    y = data[:, 1]  # Second column for y-axis
    
    # Create a scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(x, y, alpha=0.7, edgecolors='k')
    plt.title('Scatter Plot of Sample Data')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def scatter_label(data, labels):
    x = data[:, 0]  # First column for x-axis
    y = data[:, 1]  # Second column for y-axis
    
    # Create a scatter plot with colors based on labels
    plt.figure(figsize=(10, 6))
    scatter = plt.scatter(x, y, c=labels, cmap='coolwarm', alpha=0.7, edgecolors='k')
    plt.title('Scatter Plot of Sample Data')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.grid(True)
    plt.colorbar(scatter, label='Label')
    plt.show()
```

```python
from sklearn.decomposition import PCA

def pca_2d(data):
    pca = PCA(n_components=2)
    return pca.fit_transform(data)
```

```python
scatter(pca_2d(train))
```

```python
scatter_label(pca_2d(test), test_labels)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import math

def plot_distribution(df):
    # Number of columns
    num_columns = len(df.columns)
    
    # Calculate the number of rows and columns for the subplots
    num_rows = math.ceil(num_columns / 3)
    num_cols = min(num_columns, 3)
    
    # Create subplots
    fig, axes = plt.subplots(nrows=num_rows, ncols=num_cols, figsize=(15, num_rows * 5))
    
    # Flatten the axes array if there's more than one plot
    axes = axes.flatten() if num_columns > 1 else [axes]
    
    # Plot data distribution for each column
    for i, column in enumerate(df.columns):
        axes[i].hist(df[column], bins=10, edgecolor='black', alpha=0.7)
        axes[i].set_title(f'Distribution of {column}')
        axes[i].set_xlabel(column)
        axes[i].set_ylabel('Frequency')
    
    # Remove any unused axes
    for ax in axes[num_columns:]:
        ax.remove()
    
    plt.tight_layout()
    plt.show()
```
